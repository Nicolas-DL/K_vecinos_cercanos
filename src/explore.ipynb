```python
# Your code here
import pandas as pd 

df_movies = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv')
df_credits = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv')

df_movies.head()
```

```python
df_movies.to_csv('../data/raw/df_movies.csv')
df_credits.to_csv('../data/raw/df_credits.csv')
```

**Creación base de datos**

```python
import sqlite3

conn = sqlite3.connect('../data/raw/movies_db.db')
```

```python
df_movies.to_sql('tabla_movies', conn, if_exists='replace', index=False)
df_credits.to_sql('tabla_credits', conn, if_exists='replace', index=False)
```

```python
print(f'df_movies: {df_movies.columns}')
print(f'df_credits: {df_credits.columns}')
```

```python
query = """
    SELECT tc.movie_id, tm.title, tm.overview, tm.genres, tm.keywords, tc.cast, tc.crew
    FROM tabla_movies as tm
    INNER JOIN tabla_credits as tc
    ON tm.title = tc.title;
"""
df = pd.read_sql_query(query, conn)
conn.close()

df
```

**Extracción y transformación de datos json**

```python
import json

df['genres'] = df['genres'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)
df.head()
```

```python
df['keywords'] = df['keywords'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)
df['cast'] = df['cast'].apply(lambda x:[item['name'] for item in json.loads(x)][:3] if pd.notna(x) else None)
df.head()
```

```python
df['crew'] = df['crew'].apply(lambda x: next((item['name'] for item in json.loads(x) if item['job']=='Director'),None) if pd.notna(x) else None)
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x:[str(x)])
df.head()
```

```python
def quitar_espacios(lista):
    return [item.replace(' ', '') for item in lista] if isinstance(lista, list) else lista

df[['genres', 'cast', 'keywords']] = df[['genres', 'cast', 'keywords']].map(quitar_espacios)
df['crew'] = df['crew'].str.replace(' ', '')

df.head()
```

```python
def cadena_con_comas(lista):
    return ','.join(lista) if isinstance(lista, list) else ''

df['genres'] = df['genres'].apply(cadena_con_comas)
df['keywords'] = df['keywords'].apply(cadena_con_comas)
df['cast'] = df['cast'].apply(cadena_con_comas)
df['overview'] = df['overview'].apply(cadena_con_comas)

df.head()
```

```python
df['overview'] = df['overview'].str.replace(',','')
df['tags'] = df['overview'] + ' ' + df['genres'] + ' ' + df['keywords'] + ' ' + df['cast'] + ' ' + df['crew']
df['tags'] = df['tags'].str.replace(',',' ')

df['tags'][0]
```

```python
df['tags'].isnull().sum()
```

```python
df['tags'] = df['tags'].fillna('')
```

**Construcción modelo K_vecinos**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

vectorizador = TfidfVectorizer()
matriz_tfidf = vectorizador.fit_transform(df['tags'])

model = NearestNeighbors(n_neighbors=6, algorithm='brute', metric='cosine')
model.fit(matriz_tfidf)

def recomendar(pelicula):
    pelicula_minusculas = pelicula.lower() #pasar todo a minúsculas para evitar errores relacionados a minúsculas o mayúsculas
    indice_movie = df[df['title'].str.lower()==pelicula_minusculas].index[0]
    distancias, indices = model.kneighbors(matriz_tfidf[indice_movie])
    peliculas_similares = [(df['title'][i], distancias[0][j]) for j,i in enumerate(indices[0])]
    return peliculas_similares[1:]

pelicula = 'how to train your dragon'
recomendaciones = recomendar(pelicula)
for pelicula, distancia in recomendaciones:
    print('- Película {}'.format(pelicula))
```

```python
pelicula = 'el mariachi'
recomendaciones = recomendar(pelicula)
for pelicula, distancia in recomendaciones:
    print('- Película {}'.format(pelicula))
```
